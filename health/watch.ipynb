```python
import numpy as np
from datetime import datetime, timedelta
ds = datetime(year=2022, month=10, day=6, hour=17)
de = ds + timedelta(days=7)
ds, de = [pd.Timestamp(x).tz_localize(sdf["startDate"].dt.tz) for x in [ds, de]]
ds
```

```python
nsdf = sdf[(sdf["startDate"] > ds) & (sdf["endDate"] < de)]
nsdf
```

```python
mm = {"InBed": 2, "Awake": 1, "AsleepUnspecified": 0, "AsleepREM": -1, "AsleepCore": -2, "AsleepDeep": -3}
nnsdf = pd.DataFrame({
    "startDateDelta": nsdf["startDate"] - ds,
    "endDateDelta": nsdf["endDate"] - ds,
    "sourceName": nsdf["sourceName"],
    "value": nsdf["value"].map(mm)
})
nnsdf
```

```python
from matplotlib import pyplot as plt
from matplotlib import patches as mpatches

def color(name):
    name = name.lower()
    cmap = plt.get_cmap('Paired')
    if "phone" in name:
        res = cmap(1)
    elif "watch" in name:
        res = cmap(3)
    else:
        res = cmap(7)
    res = res[:3] + (0.8,)
    return res

plt.figure(figsize=(14, 7))
for row in nnsdf.itertuples():
    s, e = [x / pd.Timedelta(hours=1) for x in [row.startDateDelta, row.endDateDelta]]
    v, c = row.value, color(row.sourceName)
    plt.fill_between([s, e], [v] * 2, [v - 1] * 2, color=c)
plt.ylim([-4.2, 2.2])
plt.legend(handles=[
    mpatches.Patch(color=color("phone"), label="iPhone"),
    mpatches.Patch(color=color("watch"), label="Apple Watch"),
    mpatches.Patch(color=color("other"), label="External App")
], loc='lower left')
plt.gca().set_yticks([-3.5 + x for x in range(6)])
plt.gca().set_yticklabels(["Asleep\n(Deep)", "Asleep\n(Core)", "Asleep\n(REM)", "Asleep\n(Unknown)", "Awake", "InBed"])
plt.xlabel("Hours")
plt.title(f'Sleep Analysis ({ds} ~ {de})')
plt.tight_layout()

def check_overlap(a1, b1, a2, b2):
    return a1 <= b2 and b1 >= a2
    
durs = []
thres = 0.5 # 30min
t = thres / 2
for row in nnsdf.itertuples():
    s, e = [x / pd.Timedelta(hours=1) for x in [row.startDateDelta, row.endDateDelta]]
    v = row.value
    if v > 0:
        continue
    for i, (sx, ex) in enumerate(durs):
        if check_overlap(sx - t, ex + t, s - t, e + t):
            durs[i] = (min(s, sx), max(e, ex))
            break
    else:
        durs += [(s, e)]
print(durs)
```

```python
def plot(ds, days=7):
    de = ds + timedelta(days=days)
    ds, de = [pd.Timestamp(x).tz_localize(sdf["startDate"].dt.tz) for x in [ds, de]]
    nsdf = sdf[(sdf["startDate"] > ds) & (sdf["endDate"] < de)]
    nnsdf = pd.DataFrame({
        "startDateDelta": nsdf["startDate"] - ds,
        "endDateDelta": nsdf["endDate"] - ds,
        "sourceName": nsdf["sourceName"],
        "value": nsdf["value"].map(mm)
    })
    plt.figure(figsize=(14, 7))
    for row in nnsdf.itertuples():
        s, e = [x / pd.Timedelta(hours=1) for x in [row.startDateDelta, row.endDateDelta]]
        v, c = row.value, color(row.sourceName)
        plt.fill_between([s, e], [v] * 2, [v - 1] * 2, color=c)
    plt.ylim([-4.2, 2.2])
    plt.legend(handles=[
        mpatches.Patch(color=color("phone"), label="iPhone"),
        mpatches.Patch(color=color("watch"), label="Apple Watch"),
        mpatches.Patch(color=color("other"), label="External App")
    ], loc='lower left')
    plt.gca().set_yticks([-3.5 + x for x in range(6)])
    plt.gca().set_yticklabels(["Asleep\n(Deep)", "Asleep\n(Core)", "Asleep\n(REM)", "Asleep\n(Unknown)", "Awake", "InBed"])
    plt.xlabel("Hours")
    plt.title(f'Sleep Analysis ({ds} ~ {de})')
    plt.tight_layout()
```

```python
plot(ds=datetime(year=2021, month=11, day=2, hour=17), days=7)
```

```python
import json
res = []
for row in adf.itertuples():
    date = row.dateComponents.tz_localize("Asia/Seoul")
    date = date.value / 1e9
    res += [{
        "dateInSeconds": date,
        "activeEnergyBurnedInKcal": float(row.activeEnergyBurned),
        "activeEnergyBurnedGoalInKcal": float(row.activeEnergyBurnedGoal),
        "exerciseTimeInMinutes": row.appleExerciseTime,
        "exerciseTimeGoalInMinutes": row.appleExerciseTimeGoal,
        "standHours": row.appleStandHours,
        "standHoursGoal": row.appleStandHoursGoal,
    }]
with open('mock_activity.json', 'w', encoding='utf-8') as f:
    json.dump(res, f, indent=4)
```

```python
res = []
for row in sdf.itertuples():
    sdate, edate = row.startDate, row.endDate
    sdate, edate = sdate.value / 1e9, edate.value / 1e9
    m = {"InBed": 0, "Awake": 2, "AsleepUnspecified": 1, "AsleepREM": 5, "AsleepCore": 3, "AsleepDeep": 4}
    res += [{
        "startDateInSeconds": sdate,
        "endDateInSeconds": edate,
        "value": m[row.value],
    }]
with open('mock_sleep.json', 'w', encoding='utf-8') as f:
    json.dump(res, f, indent=4)
```

```python
htdf = pd.read_feather("heart.feather")
print(htdf.info())
htdf.describe()
res = []
prev = None
for row in htdf.itertuples():
    if prev is None:
        prev = row.startDate
    else:
        diff = row.startDate - prev
        if diff < pd.Timedelta(hours=1):
            continue
        prev = row.startDate
    date = row.startDate.value / 1e9
    res += [{
        "dateInSeconds": date,
        "valueInCountPerMinute": row.value,
    }]
with open('mock_heart.json', 'w', encoding='utf-8') as f:
    json.dump(res, f, indent=4)
```
